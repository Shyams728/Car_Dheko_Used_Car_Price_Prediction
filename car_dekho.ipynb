```python
import json
import re

# Function to convert single-quoted JSON to double-quoted JSON
def convert_to_valid_json(json_str):
    try:
        # Replace single quotes with double quotes
        json_str = re.sub(r"'", '"', json_str)
        return json.loads(json_str)
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        return None

# List to store all the combined specs
all_specs_list = []

for car_specs_data in df['new_car_specs']:
    # Try parsing if the data is a string, otherwise, proceed
    if isinstance(car_specs_data, str):
        car_specs_data = convert_to_valid_json(car_specs_data)

    # Proceed only if we have a valid dictionary after parsing
    if isinstance(car_specs_data, dict):
        # Extracting values from 'top'
        top_values = {item['key']: item['value'] for item in car_specs_data.get('top', [])}

        # Extracting values from 'data'
        specs = {}
        for section in car_specs_data.get('data', []):
            for item in section.get('list', []):
                specs[item['key']] = item['value']

        # Combine 'top' and 'data' into a single dictionary
        all_specs = {**top_values, **specs}

        # Append the combined specs to the list
        all_specs_list.append(all_specs)
    else:
        print(f"Skipping invalid data: {car_specs_data}")

# Convert the list of dictionaries into a DataFrame
df_new_car_specs = pd.DataFrame(all_specs_list)

# Display the new DataFrame
df_new_car_specs.head(2)
```

```python
"""
Cell generated by Data Wrangler.
"""
import numpy as np

def clean_data(df_new_car_specs):
    
    # Function to extract max power
    def extract_max_power(value):
        if isinstance(value, str):
            # Extract numeric value before 'bhp' or 'PS'
            match = re.search(r'(\d+\.?\d*)\s*(bhp|PS)', value)
            if match:
                return float(match.group(1))
        return None

    # Apply the function to the 'Max Power' column
    df_new_car_specs['Max_Power'] = df_new_car_specs['Max Power'].apply(extract_max_power)
    # Drop column: 'Ground Clearance Unladen'
    df_new_car_specs = df_new_car_specs.drop(columns=['Ground Clearance Unladen'])
    # Derive column 'Mileage_kmpl' from column: 'Mileage'
    # Transform based on the following examples:
    #    Mileage        Output
    # 1: "23.1 kmpl" => "23.1"
    df_new_car_specs.insert(1, "Mileage_kmpl", df_new_car_specs["Mileage"].str.split(" ").str[0])
    # Drop column: 'Mileage'
    df_new_car_specs = df_new_car_specs.drop(columns=['Mileage'])
    # Drop column: 'Engine'
    df_new_car_specs = df_new_car_specs.drop(columns=['Engine'])
    # Drop column: 'Seats'
    df_new_car_specs = df_new_car_specs.drop(columns=['Seats'])
    # Drop column: 'Values per Cylinder'
    df_new_car_specs = df_new_car_specs.drop(columns=['Values per Cylinder'])
    # Drop column: 'BoreX Stroke'
    df_new_car_specs = df_new_car_specs.drop(columns=['BoreX Stroke'])
    # Drop column: 'Seating Capacity'
    df_new_car_specs = df_new_car_specs.drop(columns=['Seating Capacity'])
    # Drop column: 'Alloy Wheel Size'
    # df_new_car_specs = df_new_car_specs.drop(columns=['Alloy Wheel Size'])
    # Replace all instances of "" with "" in column: 'Wheel Size'
    df_new_car_specs['Wheel Size'] = df_new_car_specs['Wheel Size'].str.replace("R", "", case=False, regex=False)
    # Replace missing values with the most common value of each column in: 'Wheel Size'
    df_new_car_specs = df_new_car_specs.fillna({'Wheel Size': df_new_car_specs['Wheel Size'].mode()[0]})
    # Change column type to int16 for column: 'Wheel Size'
    df_new_car_specs = df_new_car_specs.astype({'Wheel Size': 'int16'})
    # Replace missing values with the most common value of each column in: 'Color'
    df_new_car_specs = df_new_car_specs.fillna({'Color': df_new_car_specs['Color'].mode()[0]})
    # Change column type to float32 for column: 'Mileage_kmpl'
    df_new_car_specs = df_new_car_specs.astype({'Mileage_kmpl': 'float32'})
    df_new_car_specs = df_new_car_specs.astype({'No of Cylinder': 'float32'})

    def extract_numeric(value):
        if isinstance(value, str):
            # Remove any non-numeric characters except for a single decimal point
            value = re.sub(r'[^\d.]', '', value)
            # Ensure only one decimal point is present
            value = re.sub(r'\.(?=.*\.)', '', value)
            if value:
                try:
                    return float(value)  # Convert to float for standardization
                except ValueError:
                    return np.nan  # Return NaN if conversion fails
        return np.nan 

    df_new_car_specs['Torque_Nm'] = df_new_car_specs['Torque'].apply(extract_numeric)
    
    df_new_car_specs['Alloy Wheel Size'] = df_new_car_specs['Alloy Wheel Size'].apply(extract_numeric)
    df_new_car_specs['Wheel_Base'] = df_new_car_specs['Wheel Base'].apply(extract_numeric)
    df_new_car_specs['Rear_Tread'] = df_new_car_specs['Rear Tread'].apply(extract_numeric)
    df_new_car_specs['Front_Tread'] = df_new_car_specs['Front Tread'].apply(extract_numeric)
    df_new_car_specs['Kerb_Weight'] = df_new_car_specs['Kerb Weight'].apply(extract_numeric)
    df_new_car_specs['Gross_Weight'] = df_new_car_specs['Gross Weight'].apply(extract_numeric)
    df_new_car_specs['Length_mm'] = df_new_car_specs['Length'].apply(extract_numeric)
    df_new_car_specs['Width_mm'] = df_new_car_specs['Width'].apply(extract_numeric)
    df_new_car_specs['Height_mm'] = df_new_car_specs['Height'].apply(extract_numeric)
    df_new_car_specs['Turning_Radius'] = df_new_car_specs['Turning Radius'].apply(extract_numeric)
    # df_new_car_specs = df_new_car_specs.drop(columns=['Turning Radius'])
    # Replace all instances of "" with "" in column: 'Cargo Volumn'
    df_new_car_specs['Cargo Volumn'] = df_new_car_specs['Cargo Volumn'].apply(extract_numeric)
    df_new_car_specs['Top_Speed_kmph'] = df_new_car_specs['Top Speed'].apply(extract_numeric)
    df_new_car_specs['Acceleration_in_sec'] = df_new_car_specs['Acceleration'].apply(extract_numeric)
    df_new_car_specs['Gear_Box_speed'] = df_new_car_specs['Gear Box'].apply(extract_numeric)
    # df_new_car_specs['Turning_Radius_in_mtr'] = df_new_car_specs['Turning Radius'].apply(extract_numeric)
    df_new_car_specs = df_new_car_specs.drop(columns=['Length','Width','Height'])
    df_new_car_specs = df_new_car_specs.drop(columns=['Kerb Weight','Gross Weight'])
    # df_new_car_specs = df_new_car_specs.drop(columns=['Width'])
    df_new_car_specs = df_new_car_specs.drop(columns=['Torque','Max Power','Max Torque','Acceleration','Displacement'])
    df_new_car_specs = df_new_car_specs.drop(columns=['Turning Radius','Top Speed','Gear Box'])
    df_new_car_specs = df_new_car_specs.drop(columns=['Rear Tread','Front Tread','Wheel Base'])
    # 80 % data have missing values
    df_new_car_specs = df_new_car_specs.drop(columns=['Compression Ratio'])
    df_new_car_specs = df_new_car_specs.drop(columns=['Wheel Size'])
    df_new_car_specs = df_new_car_specs.fillna({'Alloy Wheel Size': df_new_car_specs['Alloy Wheel Size'].mode()[0]})
    # Replace missing values with the mode of each column in: 'No of Cylinder'
    df_new_car_specs = df_new_car_specs.fillna({'No of Cylinder': df_new_car_specs['No of Cylinder'].mode()[0]})
    # Change column type to int16 for column: 'No of Cylinder'
    df_new_car_specs = df_new_car_specs.astype({'No of Cylinder': 'int16'})
    # Replace missing values with the mean of each column in: 'Torque_Nm'
    df_new_car_specs = df_new_car_specs.fillna({'Torque_Nm': df_new_car_specs['Torque_Nm'].mean()})
    # Replace missing values with the mode of each column in: 'Wheel_Base'
    df_new_car_specs = df_new_car_specs.fillna({'Wheel_Base': df_new_car_specs['Wheel_Base'].mode()[0]})
    # Change column type to int16 for column: 'Wheel_Base'
    df_new_car_specs = df_new_car_specs.astype({'Wheel_Base': 'int16'})
    # Replace missing values with the mean of each column in: 'Mileage_kmpl'
    df_new_car_specs = df_new_car_specs.fillna({'Mileage_kmpl': df_new_car_specs['Mileage_kmpl'].mean()})
    # Replace missing values with "not available" in column: 'Engine Type'
    df_new_car_specs = df_new_car_specs.fillna({'Engine Type': "not available"})
    # Replace missing values with "default" in column: 'Engine Type'
    df_new_car_specs = df_new_car_specs.fillna({'Engine Type': "default"})
    # Replace missing values with "NA" in column: 'Value Configuration'
    df_new_car_specs = df_new_car_specs.fillna({'Value Configuration': "NA"})
    # Replace missing values with "NA" in column: 'Fuel Suppy System'
    df_new_car_specs = df_new_car_specs.fillna({'Fuel Suppy System': "NA"})
    # Replace missing values with "NA" in column: 'Turbo Charger'
    df_new_car_specs = df_new_car_specs.fillna({'Turbo Charger': "NA"})
    # Replace missing values with "NA" in column: 'Super Charger'
    df_new_car_specs = df_new_car_specs.fillna({'Super Charger': "NA"})
    # Replace gaps forward from the previous valid value in: 'Drive Type'
    df_new_car_specs = df_new_car_specs.fillna({'Drive Type': df_new_car_specs['Drive Type'].ffill()})
    # Replace missing values with the most common value of each column in: 'Steering Type'
    df_new_car_specs = df_new_car_specs.fillna({'Steering Type': df_new_car_specs['Steering Type'].mode()[0]})
    # Replace gaps forward from the previous valid value in: 'Front Brake Type', 'Rear Brake Type', 'Tyre Type'
    df_new_car_specs = df_new_car_specs.fillna({'Front Brake Type': df_new_car_specs['Front Brake Type'].ffill(), 'Rear Brake Type': df_new_car_specs['Rear Brake Type'].ffill(), 'Tyre Type': df_new_car_specs['Tyre Type'].ffill()})
    # Replace missing values with the most common value of each column in: 'No Door Numbers'
    df_new_car_specs = df_new_car_specs.fillna({'No Door Numbers': df_new_car_specs['No Door Numbers'].mode()[0]})
    # Change column type to int8 for column: 'No Door Numbers'
    df_new_car_specs = df_new_car_specs.astype({'No Door Numbers': 'int8'})
    # Replace missing values with the median of each column in: 'Cargo Volumn'
    df_new_car_specs = df_new_car_specs.fillna({'Cargo Volumn': df_new_car_specs['Cargo Volumn'].median()})
    # Replace missing values with the median of each column in: 'Max_Power'
    df_new_car_specs = df_new_car_specs.fillna({'Max_Power': df_new_car_specs['Max_Power'].median()})
    # Replace missing values with the median of each column in: 'Front_Tread', 'Rear_Tread' and 2 other columns
    df_new_car_specs = df_new_car_specs.fillna({'Front_Tread': df_new_car_specs['Front_Tread'].median(), 'Rear_Tread': df_new_car_specs['Rear_Tread'].median(), 'Kerb_Weight': df_new_car_specs['Kerb_Weight'].median(), 'Gross_Weight': df_new_car_specs['Gross_Weight'].median()})
    # Replace missing values with the median of each column in: 'Width_mm', 'Length_mm', 'Height_mm'
    df_new_car_specs = df_new_car_specs.fillna({'Width_mm': df_new_car_specs['Width_mm'].median(), 'Length_mm': df_new_car_specs['Length_mm'].median(), 'Height_mm': df_new_car_specs['Height_mm'].median()})
    # Change column type to int16 for columns: 'Length_mm', 'Width_mm' and 3 other columns
    df_new_car_specs = df_new_car_specs.astype({'Length_mm': 'int16', 'Width_mm': 'int16', 'Height_mm': 'int16', 'Gross_Weight': 'int16', 'Kerb_Weight': 'int16'})
    # Replace missing values with the mode of each column in: 'Turning_Radius'
    df_new_car_specs = df_new_car_specs.fillna({'Turning_Radius': df_new_car_specs['Turning_Radius'].mode()[0]})
    # Replace missing values with the mode of each column in: 'Gear_Box_speed'
    df_new_car_specs = df_new_car_specs.fillna({'Gear_Box_speed': df_new_car_specs['Gear_Box_speed'].mode()[0]})
    # Change column type to int8 for column: 'Gear_Box_speed'
    df_new_car_specs = df_new_car_specs.astype({'Gear_Box_speed': 'int8'})
    # Replace missing values with the median of each column in: 'Gear_Box_speed'
    df_new_car_specs = df_new_car_specs.fillna({'Gear_Box_speed': df_new_car_specs['Gear_Box_speed'].median()})
    # Replace missing values with the median of each column in: 'Acceleration_in_sec', 'Top_Speed_kmph'
    df_new_car_specs = df_new_car_specs.fillna({'Acceleration_in_sec': df_new_car_specs['Acceleration_in_sec'].median(), 'Top_Speed_kmph': df_new_car_specs['Top_Speed_kmph'].median()})
    
    return df_new_car_specs

df_new_car_specs_clean = clean_data(df_new_car_specs.copy())
df_new_car_specs_clean.head()
```
